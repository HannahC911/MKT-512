```python
import pandas as pd

url = 'https://raw.githubusercontent.com/HannahC911/mkt512challenge2/main/crews-cup.csv'
data = pd.read_csv(url)
data
```

```python
import os

import numpy as np
import seaborn

# Assigning the list of columns from the dataset
numericColumns = ['a1','a2','a3','n1','n2','n3','e1','e2','e3','f1','f2','f3','h1','h2','h3','csat','intent','origination','classes','profit']

# Creating a correlation matrix
correlationMatrix  = data.loc[:,numericColumns].corr()
# Printing the correlation matrix.
print(correlationMatrix)
# Displaying the correlation matrix
seaborn.heatmap(correlationMatrix, annot=False)
```

```python
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
```

```python
data.drop(['id','csat','intent','origination','classes','profit'],axis=1,inplace=True)
```

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(data)

kmo_model
```

```python
fa = FactorAnalyzer(rotation='varimax')
fa.fit(data)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev
```

```python
# Create scree plot using matplotlib
plt.scatter(range(1,data.shape[1]+1),ev)
plt.plot(range(1,data.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

```python
fa = FactorAnalyzer(n_factors=5,rotation='varimax')
fa.fit(data)
```

```python
factor_df=pd.DataFrame(fa.loadings_,columns=['Factor1','Factor2','Factor3','Factor4','Factor5'],index=data.columns)
print(factor_df)
```

```python
fa.diagram(data)
```

```python
seaborn.heatmap(factor_df, annot=True)
plt.show()
```

```python
import statsmodels.api as sm
```

```python
url2 = 'https://raw.githubusercontent.com/HannahC911/mkt512challenge2/main/crews-cup2.csv'
datareg = pd.read_csv(url2)
datareg

```

```python
x = datareg[['a_avg','n_avg','e_avg','f_avg','h_avg']]
y = datareg['intent']

x = sm.add_constant(x)  # Add a constant term to the model
model = sm.OLS(y, x).fit()  # Ordinary Least Squares (OLS) regression
print(model.summary())
```
